```python
import pandas as pd
```

```python
userdata = {
    'name': ['Bhavesh','Sandesh','Bhagyashri'],
    'age': [20, 21, 22],
    'salary': [50000, 60000, 70000],
    'gender':['Male','Male','Female']
}
user_df = pd.DataFrame(userdata)
print(user_df)
```

```python
print(f'second index is : {user_df.iloc[1]}')
```

```python
first_username = user_df.iloc[0]['salary']
print(first_username)
```

```python
first_username = user_df.iloc[0]
print(first_username['name'])
```

```python
first_index_0 = user_df.loc[0]
print(first_index_0)
```

```python
sales_data = [
    {
        'product': 'Laptop',
        'price': 1000,
        'quantity': 5
    },
    {
        'product': 'Smartphone',
        'price': 500,
        'quantity': 10
    },
    {
        'product': 'Tablet',
        'price': 300,
        'quantity': 7
    }]
```

```python
sales_df = pd.DataFrame(sales_data,index = ['d','r','t'])
print(sales_df)
```

```python
""" 
while iloc is used to access a group of rows and columns by integer position.
"""
sales_df_loc = sales_df.iloc[0] 
print(sales_df_loc)
```

```python
"""
loc is used to access a group of rows and columns by labels or a boolean array,
"""
sales_df_loc = sales_df.loc['r'] #loc is used to access a group of rows and columns by labels or a boolean array, while iloc is used to access a group of rows and columns by integer position.
print(sales_df_loc)
```

```python
print(sales_df['product'])
```

```python
col_name = ['product','quantity']
names = sales_df[col_name]
print(names)
```

```python
col_name = ['product','quantity']
names = sales_df[col_name].iloc[0]
print(names)
```
