### While Loop Example

```python
num = 5

while num > 0:
    print(num)
    num -= 1
```

```python
num = 1

while True:
    if num == 4:
        break
    print(num)
    num += 1
```

```python
num = 0

while num < 7:
    num += 1
    if num % 2 == 0:
        continue
    print(num)
```

```python
count = 1

while count <= 3:
    if count == 2:
        pass  # logic to be added later
    print(count)
    count += 1
```

```python
password = ""

while password != "admin123":
    password = input("Enter the password: ")  # assume user entered correct password

    if password == "admin123":
        print("correct password")
    else:
        print("Invalid! Please Enter correct password")
    
```

```python
num = 1
total = 0

while num <= 5:
    total += num
    num += 1

print("Sum:", total)
```

### for loop

```python
for i in range(2, 11, 2):
    print(i)
```

```python
for num in range(1, 10):
    if num == 6:
        break
    print(num)
```

```python
for num in range(1, 10):
    if num % 3 == 0:
        continue
    print(num)
```

```python
for i in range(1, 4):
    if i == 2:
        pass  
    print(i)
```

```python
for row in range(1, 4):
    for col in range(1, 3):
        print(f"Row {row}, Col {col}")
```

```python
word = "PYTHON"

for char in word:
    print(char)
```

```python
student = {"name": "Alice", "age": 21, "course": "CS"}

for value in student:
    print(value)
```

```python
marks = {"Math": 85, "Physics": 90, "Chemistry": 88}

for value in marks.values():
    print(value)
```

```python
employee = {"id": 101, "name": "John", "salary": 50000}

for key, value in employee.items():
    print(key, ":", value)
```

```python
scores = {"A": 90, "B": 45, "C": 78}

for grade, score in scores.items():
    if score < 50:
        continue
    print(grade, score)
```

```python
products = {"pen": 10, "book": 50, "laptop": 50000}

for item, price in products.items():
    if price > 1000:
        break
    print(item, price)
```
