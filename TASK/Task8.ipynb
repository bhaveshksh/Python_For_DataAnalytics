```python
sent1 = "A man, a plan, a canal: Panama"
```

```python
def is_palindrome_while(sentence):
    cleaned = ""
    for ch in sentence.lower():
        if ch.isalnum():
            cleaned += ch

    left = 0
    right = len(cleaned) - 1

    while left < right:
        if cleaned[left] != cleaned[right]:
            return False
        left += 1
        right -= 1

    return True


print(is_palindrome_while(sent1))
```
