#### 1st if-elif-else example by using string

````python
str = 'goody'
if str == 'gody':
    print("Wrong word")
elif str == 'goody':
    print("Right Word")
else:
    print("Please Right Correct word")
````

#### 2nd if-elif-else example by using number

````python
num = int(input("Enter the number: "))

if num > 100:
    print(f"{num} is less than 100")
elif num < 80:
    print(f"{num} is less than 80")
elif num >= 50:
    print(f"{num} is greater than equal to 50")
else:
    print(f"please enter num")
````

#### 3rd if-elif-else example by using list

````python
general = ['Math', 'chemistry', 'physics','Biology']
if general in general:
    print("This Student is all for all type of example like JEE and NEET")
elif general == general.pop(3):
    print("This Student is allow for JEE")
else:
    print("This student is not allow for exam")
````

### Dict

````python
d={'name':'Alice', 'age':30}
if d['age'] < 18:
    print(f"{d['name']} is a minor.")
elif d['age'] < 65:
    print(f"{d['name']} is a adult.")
else:
    print(f"{d['name']} is a senior.")
````

### Set

````python
my_set = {'x', 'y', 'z'}
value = input()

if value in my_set:
    if value == 'x':
        print("Value is x")
    elif value == 'y':
        print("Value is y")
    else:
        print("Value is z")
else:
    print("Not in the set")
````

### Tuple

````python
colors = ("red", "green", "blue")
color = "green"

if color == colors[0]:
    print("Color is red")
elif color == colors[1]:
    print("Color is green")
else:
    print("Color is blue")
````

## String

````python
word = "goody"

match word:
    case "gody":
        print("Wrong word")
    case "goody":
        print("Right word")
    case _:
        print("Unknown word")
````

### number

````python
num = 2

match num:
    case 1:
        print("One")
    case 2:
        print("Two")
    case 3:
        print("Three")
    case _:
        print("Invalid number")
````

### List

````python
subjects = ["Math", "Physics"]

match subjects:
    case ["Math", "Physics"]:
        print("Engineering stream")
    case ["Biology", "Chemistry"]:
        print("Medical stream")
    case _:
        print("Other stream")
````

### set

````python
my_set = {"x", "y", "z"}

match my_set:
    case s if s == {"x", "y", "z"}:
        print("Set contains x, y, z")
    case s if s == {"x", "y"}:
        print("Set contains x and y")
    case _:
        print("Different set")
````

### Tuple

````python
point = (0, 1)

match point:
    case (0, 0):
        print("Origin")
    case (0, y):
        print(f"On Y-axis at {y}")
    case (x, 0):
        print(f"On X-axis at {x}")
    case _:
        print("Somewhere else")
````
