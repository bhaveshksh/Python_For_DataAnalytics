create a class name TourPlanner Tour Planner class must have init method to intialize Name of Company provide a list of dict of Tour package in init method and define global instance variable of tour_packages { 'name' : 'tour name' , price : 200 } create a method named book will contain 3 parameter person_name, number_of_people and package_name create a method to provide bill of package based on what person booked , bill will based on how many number of person was there basic code fore begineer in python


```python
class TourPlanner:
    def __init__(self, company_name):
        # Initialize company name
        self.company_name = company_name
        
        # List of tour packages (global instance variable)
        self.tour_packages = [
            {'name': 'Beach Holiday', 'price': 200},
            {'name': 'Mountain Trek', 'price': 300},
            {'name': 'City Tour', 'price': 150}
        ]
        
        # To store bookings
        self.bookings = []

    def book(self, person_name, number_of_people, package_name):
        # Find the package by name
        for package in self.tour_packages:
            if package['name'] == package_name:
                # Save booking details
                booking = {
                    'person_name': person_name,
                    'package_name': package_name,
                    'number_of_people': number_of_people,
                    'price_per_person': package['price']
                }
                self.bookings.append(booking)
                print(f"Booking successful for {person_name} on {package_name}!")
                return
        print("Package not found!")

    def generate_bill(self, person_name):
        # Find booking for the person
        for booking in self.bookings:
            if booking['person_name'] == person_name:
                total_bill = booking['number_of_people'] * booking['price_per_person']
                print(f"Bill for {person_name}: ${total_bill}")
                return total_bill
        print("No booking found for this person.")
        return None


# Example usage:
planner = TourPlanner("Happy Travels")

planner.book("Alice", 3, "Beach Holiday")
planner.book("Bob", 2, "Mountain Trek")

planner.generate_bill("Alice")
planner.generate_bill("Bob")
```
