### create a multiLevel class example of 4 class of case study hospital mangment

```python
class College:

  def __init__(self,cName,cLocation,cId):
    self.c_name = cName
    self.c_location = cLocation
    self.c_id = cId

  def get_college_id(self):
    return self.c_id

  def get_college_details(self):
    return {
        'college_name': self.c_name,
        'college_location': self.c_location,
        'college_id': self.c_id
    }


class Department(College):

  def __init__(self,dId,dName,dStaff,cName,cLocation,cId):
    super().__init__(cName,cLocation,cId)
    self.d_id = dId
    self.d_name = dName
    self.d_staff = dStaff

  def get_department_details(self):
    return {
        'department_id': self.d_id,
        'department_name': self.d_name,
        'department_staff': self.d_staff
    }




d1 = Department('d_101','Data Analytics',4,'ARC','Nandanvan','c_101')
d1_college_id = d1.get_college_id()
print('d1_college_id',d1_college_id)
d1_college_details = d1.get_college_details()
print('d1_college_details',d1_college_details)
d1_department_details = d1.get_department_details()
print('d1_department_details',d1_department_details)
```
