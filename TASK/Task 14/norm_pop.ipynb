```python
import pandas as pd
```

```python
pop_10_19_clean = pd.read_csv('./clean_datasets/population-2010-2019.csv')
pop_20_29_clean = pd.read_csv('./clean_datasets/population-2020-2029.csv')
pop_30_39_clean = pd.read_csv('./clean_datasets/population-2030-2040.csv')
```

```python
pop_10_19_clean.head()
```

```python
pop_30_39_clean.head()
```

```python
all_pop = pd.concat([pop_10_19_clean, pop_20_29_clean, pop_30_39_clean], ignore_index=True)
```

```python
all_pop.head()
```

```python
all_pop.info()
```

```python
# Example: Pivot so each row is AgeGrp+Time, columns are PopMale and PopFemale
pivot_df = all_pop.pivot_table(
    index=['AgeGrp', 'Year'], 
    values=['PopMale', 'PopFemale'],
    aggfunc='sum'  # or 'mean', or another aggregation if needed
)

pivot_df
```

```python
all_pop_data = all_pop.melt(
    id_vars=[col for col in all_pop.columns if col not in ['PopMale', 'PopFemale']],
    value_vars=['PopMale', 'PopFemale'],
    var_name='Gender',
    value_name='Population'
)

all_pop_data
```

```python
all_pop_data.info()
```

```python
all_pop_data['Gender'].replace({'PopMale': 'Male', 'PopFemale': 'Female'}, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler

# normalize the population column using min-max scaling
scaler = MinMaxScaler()
all_pop_data['Population_norm'] = scaler.fit_transform(all_pop_data[['Population']])

# alternatively, z-score normalization could be added:
# all_pop_data['Population_zscore'] = (
#     all_pop_data['Population'] - all_pop_data['Population'].mean()
# ) / all_pop_data['Population'].std()

all_pop_data.head()
```
