calcuate PopMale by substracting PopTotal - PopFemale and repalce value where pop_30_39['PopMale'] == 'ERROR_6.246'


```python
# Replace 'ERROR_6.246' in PopMale with PopTotal - PopFemale
mask = pop_30_39['PopMale'] == 'ERROR_6.246'
mask
pop_30_39.loc[mask, 'PopMale'] = pop_30_39.loc[mask, 'PopTotal'] - pop_30_39.loc[mask, 'PopFemale']
```

```python
pop_30_39['PopMale'] = pop_30_39['PopMale'].astype(float)
```
