````python
from abc import ABC, abstractmethod 
class Animal(ABC):
    #@abstractmethod
    def make_sound(self):
        print("Meow")

class Dog(Animal):
    def make_sound(self):
        print("Bark")
````

````python
class A:
    def greet(self):
        print("Hello from A")

class B:
    def greet(self):
        print("Hello from B")

class C(A, B):
    def greet(self):
        super().greet()

c = C()	
c.greet()  
print(C.__mro__)
c.greet()
````

````python
def apply_twice(func, value):
    return func(value)

def square(x):
    return x * x

print(apply_twice(square, 2))  
````

````python
def decorator(func):
    def wrapper():
        print("Before function call")
        func()
        print("After function call")
    return wrapper
@decorator
def greet():
    print("Hello!")

greet()
````

````python
import copy
original = [[1, 2], [3, 4]]
shallow = copy.copy(original)
print(shallow)
shallow[0][0] = 2
print(shallow)
deep = copy.deepcopy(original)
original[0][0] = 99
print(original)
print(shallow[0][0])  # 99 (affected)
print(deep[0][0])     # 1 (not affected)
print(shallow)
````
