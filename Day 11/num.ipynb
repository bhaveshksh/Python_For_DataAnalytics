```python
import numpy as np
```

```python
np1 = np.arange(1,11)

print(np1 + 1)
print(np1 * 2)
print(np1 - 3)
print(np1 / 4)
```

```python
np2 = list(range(1,11))

np3 = []
for i in np2:
    print(i*2)
    np3.append(i)
```
